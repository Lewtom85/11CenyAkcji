Pobieranie bilbiotek

```python
import pandas as pd
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
import cufflinks as cf
cf.go_offline()
import plotly.offline as pyo
import plotly.graph_objs as go
from plotly import subplots

miedz = pd.read_csv('ca_c_f_d.csv',parse_dates=True)
kghm = pd.read_csv('kgh_d.csv', parse_dates=True)
```

```python
miedz.head(5)
```

```python
kghm.head(5)
```

Stworzenie dwoch wykresow liniowych

```python
fig = subplots.make_subplots(rows=2,cols=1,subplot_titles=['KGHM','Mied'],shared_xaxes=True)
fig.add_trace(go.Scatter(x=kghm['Data'],
                         y=kghm['Zamkniecie'], name='KGHM'),1,1)

fig.add_trace(go.Scatter(x=miedz['Data'],
                         y=miedz['Zamkniecie'], name='Mied'),2,1)
fig.update_layout(title='Ceny zamknicia wedug daty', title_x=0.5)
iplot(fig)
```

Tworzenie tabeli i dodwanie jej do figury:

```python
fig = go.Figure()

table = go.Table(
    header = dict(values = ["Data", "KGHM", "Mied"],
                                    align = 'left' ),
    cells = dict(values = [kghm['Data'], kghm['Zamkniecie'], miedz['Zamkniecie']],
                                  align = 'left')
)
fig.update_layout(title='Ceny zamknicia wedug daty',title_x=0.5)
fig.add_trace(table)
fig.show()
```
